```python
import random 
import matplotlib.pyplot as plt
from PIL import Image
import numpy as np
```

```python
x,y = 2,3
fig, axs = plt.subplots(x,y, figsize=(12, 4))
plt.subplots_adjust(wspace=0.00, hspace=0.0)

# Replace the next line with your actual list of image paths
regions = ["Europe", "Africa"]
obj = "bag"
folders = ["./geode/", "./generated/", "./generated-new/" ]
image_paths = []
for region in regions:
    for folder in folders:
        image_paths.append(folder + obj + "-" + region + ".png")
for i,_ in enumerate(axs.flat):
    img = Image.open(image_paths[i])
    idx = i
    axs[idx // y,idx%y].imshow(np.asarray(img))
    axs[idx // y,idx%y].set_xticks([], [])
    axs[idx // y,idx%y].set_yticks([], [])

    axs[idx // y,idx%y].spines['top'].set_visible(False)
    axs[idx // y,idx%y].spines['right'].set_visible(False)
    axs[idx // y,idx%y].spines['bottom'].set_visible(False)
    axs[idx // y,idx%y].spines['left'].set_visible(False)
axs[0,0].set_ylabel(f'{regions[0]}', fontsize=25, labelpad=10)
axs[1,0].set_ylabel(f'{regions[1]}', fontsize=25, labelpad=10)

axs[0, 0].set_title(f"GeoDE", fontsize=25, pad=10)
# large_font = r"\fontsize{25}{30}\selectfont"  # Customize the large font size as needed
# small_font = r"\fontsize{15}{18}\selectfont"
# title_text = f"{large_font}LDM\n{small_font}object in region " 
axs[0, 1].set_title(f"LDM (strategy-1)", fontsize=25, pad=10)
axs[0, 2].set_title(f"LDM (strategy-2)", fontsize=25, pad=10)


# title_obj.set_position([1.00, 1.03])

plt.savefig(f"main_obj-" + obj + ".pdf",bbox_inches='tight', format='pdf', dpi=400)
plt.show()
```
