```python
import numpy as np
import pandas as pd
import pickle
import matplotlib.pyplot as plt
from mpl_toolkits.axes_grid1 import ImageGrid
import matplotlib.image as mpimg
import torchvision.transforms as transforms
import sys
sys.path.append("/home/jupyter/DIG-In/src/vit_patching/models")
import patch_utils as pu
np.random.seed(42)
import get_patch_ViT16_features_v1 as gp
from lang_sam import LangSAM
from PIL import Image
import torch
import copy
import imp
imp.reload(pu)
import torchvision.transforms.functional as F
import json
```

```python
img_size = 224
n_patches = 224 // 16
transforms_list = transforms.Compose([
        gp.CenterCropLongEdge(),
        transforms.Resize(img_size),  # Use dynamic img_size based on the model
        transforms.ToTensor(),
        # TRANSFORM_NORMALIZE
    ])
    
preprocessor1 = transforms_list
preprocessor2 = gp.TRANSFORM_NORMALIZE
```

```python
model = LangSAM()
```

```python
full_img_root = "/home/jupyter/geode/images"
seg_obj_root = full_img_root
seg_bg_root = full_img_root
full_img_gen_root = "/home/jupyter/696DS-data/generated-data/LDM-1-5/geode_prompts_regions_balanceddataset"
seg_obj_gen_root = full_img_gen_root
seg_bg_gen_root = full_img_gen_root
images_folders = [(full_img_root, full_img_gen_root), (seg_obj_root, seg_obj_gen_root), (seg_bg_root, seg_bg_gen_root)]
```

```python
analysis_file = "/home/jupyter/DIG-In/src/final/manifold_qualitative_sample1.json"
```

```python
with open(analysis_file, 'r') as f:
    manifold_Images = json.load(f)
```

## TTF: failure mode - 1 

```python
setting = manifold_Images['True_True_False'][0]
setting
```

```python
# TTF

real_image = f"{full_img_root}/{setting['img_id']}"
fake_full_mfld = [f"{full_img_gen_root}/{x}" for x in setting['content']['fakeImages'][0]]
fake_obj_mfld = [f"{full_img_gen_root}/{x}" for x in setting['content']['fakeImages'][1]]
```

```python
full_setting = "Full"
full_title = "GeoDE"
gen_full_title = "LDM"
obj_setting = "Obj-only"
obj="car"
x,y = 2,3
fig, axs = plt.subplots(x,y, figsize=(9,6))
padding_value = 0.02
plt.subplots_adjust(wspace=0.02, hspace=0.02)
img_paths = [real_image] + fake_full_mfld + [real_image] + fake_obj_mfld[:2]
for i in range(len(img_paths)):
    image_pil = Image.open(img_paths[i])
    img2 = preprocessor1(image_pil)
    
    if i < 3:
        img3 = F.to_pil_image(img2)
    else:
        masks, boxes, phrases, logits = model.predict(image_pil, obj)

        merged_tensor = torch.any(masks, dim=0, keepdim=True)
        numpy_mask = merged_tensor.squeeze().numpy()
        img, patch_mask = gp.relevant_image_and_patch_mask(image_pil, numpy_mask, 224 //16, preprocessor1, do_obj_seg=False, use_compliment_for_BG=True, use_blackened=False)

        img3 = pu.border_and_alpha(img, patch_mask, fill_alpha=True, alpha_filler=200)
        
    
    axs[i//y,i%y].imshow(img3)
    axs[i//y,i%y].set_xticks([], [])
    axs[i//y,i%y].set_yticks([], [])


    axs[i//y,i%y].spines['top'].set_visible(False)
    axs[i//y,i%y].spines['right'].set_visible(False)
    axs[i//y,i%y].spines['bottom'].set_visible(False)
    axs[i//y,i%y].spines['left'].set_visible(False)

axs[0,0].set_position([axs[0,0].get_position().x0- padding_value, axs[0,0].get_position().y0, 
                             axs[0,0].get_position().width , axs[0,0].get_position().height])
axs[0,0].set_ylabel(f"{full_setting}", fontsize=25)
axs[0,0].set_title(f"{full_title}", fontsize=25)
axs[1,0].set_position([axs[1,0].get_position().x0- padding_value, axs[1,0].get_position().y0, 
                             axs[1,0].get_position().width , axs[1,0].get_position().height])
# obj setting
axs[1,0].set_ylabel(f"{obj_setting}", fontsize=25)

title_obj = axs[0,1].set_title(f"{gen_full_title}", fontsize=25)
title_obj.set_position([1.00, 1.03])
plt.savefig(f"ldm_failure_modes/ttf_" + obj + ".pdf",bbox_inches='tight', format='pdf', dpi=400)

plt.show()
```
