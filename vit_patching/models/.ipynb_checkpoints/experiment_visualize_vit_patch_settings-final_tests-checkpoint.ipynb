```python
import get_patch_ViT16_features_v1 as gp
import torchvision.transforms as transforms
from lang_sam import LangSAM
import numpy as np
import pandas as pd
from PIL import Image
import torch
import copy
```

```python
model = LangSAM()
```

```python
import imp
imp.reload(gp)
```

```python
img_size = 224
n_patches = 224 // 16
transforms_list = transforms.Compose([
        gp.CenterCropLongEdge(),
        transforms.Resize(img_size),  # Use dynamic img_size based on the model
        transforms.ToTensor(),
        # TRANSFORM_NORMALIZE
    ])
    
preprocessor1 = transforms_list
preprocessor2 = gp.TRANSFORM_NORMALIZE
```

```python
img_dir = "/work/pi_dhruveshpate_umass_edu/project_21/data/geode/images"
df_path = "/home/asureddy_umass_edu/DIG-In/data/geode_27_objects.csv"
df = pd.read_csv(df_path)
img_path = f"{img_dir}/{df.iloc[0,:]['file_path']}"
print(img_path)
obj = "bag"
img_path
image_pil = Image.open(img_path)
```

```python
image_pil
```

```python
masks, boxes, phrases, logits = model.predict(image_pil, obj)
```

```python
merged_tensor = torch.any(masks, dim=0, keepdim=True)
numpy_mask = merged_tensor.squeeze().numpy()
```

```python
# object segmented images for different patch vit settings
# effectively only 2 settings for obj seg case
settings = [(False, False), (False, True), (True, False), (True, True)]
for setting in settings:
    print(f"use_compliment_for_BG: {setting[0]}, use_blackened: {setting[1]}")
    img, patch_mask = gp.relevant_image_and_patch_mask(image_pil, numpy_mask, 224 //16, preprocessor1, do_obj_seg=True, use_compliment_for_BG=setting[0], use_blackened=setting[1])
    img2 = copy.deepcopy(img)
    img3 = gp.mask_image(img2, patch_mask[0], filler_val=0.99)
    display(transforms.ToPILImage()(img3))
```

```python
# BG segmented images for different patch vit settings
# effectively only 3 settings for obj seg case
settings = [(False, False), (False, True), (True, False), (True, True)]
for setting in settings:
    print(f"use_compliment_for_BG: {setting[0]}, use_blackened: {setting[1]}")
    img, patch_mask = gp.relevant_image_and_patch_mask(image_pil, numpy_mask, 224 //16, preprocessor1, do_obj_seg=False, use_compliment_for_BG=setting[0], use_blackened=setting[1])
    img2 = copy.deepcopy(img)
    img3 = gp.mask_image(img2, patch_mask[0], filler_val=0.99)
    display(transforms.ToPILImage()(img3))
```
